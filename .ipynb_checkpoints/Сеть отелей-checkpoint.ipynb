# Прогнозирование оттока клиентов в сети отелей «Как в гостях»

### Шаг 0. Введение в проект

Заказчик этого исследования — сеть отелей «Как в гостях». Чтобы привлечь клиентов, эта сеть отелей добавила на свой сайт возможность забронировать номер без предоплаты. Однако если клиент отменял бронирование, то компания терпела убытки. Сотрудники отеля могли, например, закупить продукты к приезду гостя или просто не успеть найти другого клиента. Чтобы решить эту проблему, мне **необходимо разработать систему, которая предсказывает отказ от брони. Если модель покажет, что бронь будет отменена, то клиенту предлагается внести депозит. Размер депозита — 80% от стоимости номера за одни сутки и затрат на разовую уборку.** Деньги будут списаны со счёта клиента, если он всё же отменит бронь.

Для разработки модели были предоставлены две выборки: train и test. В каждой из которых **следующие столбцы:**

1. id — номер записи;
2. adults — количество взрослых постояльцев;
3. arrival_date_year — год заезда;
4. arrival_date_month — месяц заезда;
5. arrival_date_week_number — неделя заезда;
6. arrival_date_day_of_month — день заезда;
7. babies — количество младенцев;
8. booking_changes — количество изменений параметров заказа;
9. children — количество детей от 3 до 14 лет;
10. country — гражданство постояльца;
11. customer_type — тип заказчика:
- Contract — договор с юридическим лицом;
- Group — групповой заезд;
- Transient — не связано с договором или групповым заездом;
- Transient-party — не связано с договором или групповым заездом, но связано с бронированием типа Transient.
12. days_in_waiting_list — сколько дней заказ ожидал подтверждения;
13. distribution_channel — канал дистрибуции заказа;
14. is_canceled — отмена заказа;
15. is_repeated_guest — признак того, что гость бронирует номер второй раз;
16. lead_time — количество дней между датой бронирования и датой прибытия;
17. meal — опции заказа:
- SC — нет дополнительных опций;
- BB — включён завтрак;
- HB — включён завтрак и обед;
- FB — включён завтрак, обед и ужин.
18. previous_bookings_not_canceled — количество подтверждённых заказов у клиента;
19. previous_cancellations — количество отменённых заказов у клиента;
20. required_car_parking_spaces — необходимость места для автомобиля;
21. reserved_room_type — тип забронированной комнаты;
22. stays_in_weekend_nights — количество ночей в выходные дни;
23. stays_in_week_nights — количество ночей в будние дни;
24. total_nights — общее количество ночей;
25. total_of_special_requests — количество специальных отметок.

```python
# Пропишем необходимые функции

def data_info(df):
    '''Информации о DF. Методы: describe, info, с просмотром памяти'''
    print(f'Размеры DF: {df.shape}', '\n')
    display(df.head(10), df.describe(), df.info(memory_usage='deep'))

def data_spread(column_list, df):
    '''Разброс данных. Отображение диаграммы размаха и гистограммы. Входные данные list колонок и df'''
    for column in column_list:
        df.boxplot(column=column)
        plt.title(f'Диаграмма размаха на столбце {column}')
        plt.show()
        df[column].plot(kind='hist', bins=100, grid=True)
        plt.title(f'Гистограмма на столбце {column}')
        plt.xlabel(column)
        plt.ylabel('Кол-во наблюдений')
        plt.show()
        print('--------------------------------------------')

def uniqueness(colomn_list, df):
    '''Просмотр уникальных значений листа'''
    for column in colomn_list:
        print(f'Количество уникальных значений столбца {column}: {df[column].nunique()}')
        print(df[column].unique())
        print('-----------------------------')
        
def splitter(data, target):
    '''Разбивает DF на признаки и таргет. На входе: DF и target, на выходе: фичи и целевой'''
    features = data.drop(target, axis=1)
    target = data[target]
    return features, target
    
def removing_outliers(data, column):
    '''Удаляет выбросы в выбранном столбце, основан на boxplot, оставляет 99% данных. На входе: DF и необходимый признак, на выходе готовый признак'''
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = (q3 - q1) * 1.5
    return data[(data[column] > q1-iqr) & (data[column] < q3+iqr)]

def violentplot(df, column_list, target):
    for column in column_list:
        plt.figure(figsize = (12, 10))
        sns.violinplot(
            x = target,
            y= column,
            data = df,
            kind = "violin",
            split = True,
            palette = "coolwarm"
        )
        plt.show()
        
        if df[column].nunique() <= 25:
            plt.figure(figsize = (12, 10))
            sns.histplot(data=df,
                         x=column,
                         binwidth=1,
                         hue=target)
            plt.show()

        print('Корреляция = ', round(df[column].corr(df[target]), 2))
        print('-----------------------------------------------------------------------------------------------')
        print()
        
def upsampling(features, target, repeat):
    '''Увеличение выборки. Техника upsampling. Вначале возвращает особенности, а затем target'''
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    features_upsampled = pd.concat([features_zeros] +  [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

def downsampling(features, target, fraction):
    '''Уменьшение выборки. Техника downsampling. Вначале возвращает особенности, а затем target'''
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    features_downsampled = pd.concat([features_zeros.sample(frac = fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat([target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled
```

```python
# Для общей аналитики совместим две выборки воедино
data = train.merge(test, how='outer')
data_info(data)
```

```python
# Посмотрим псещаемость предприятия

# Создадим словарь для месяцев
month_list = {'January' : 1,
              'February' : 2,
              'March' : 3,
              'April' : 4,
              'May' : 5,
              'June' : 6,
              'July' : 7,
              'August' : 8,
              'September' : 9,
              'October' : 10,
              'November' : 11,
              'December' : 12} 

# Создадим своднуб таблицу по году, месяцу и кол-ву посетителей
data_attendance = data.query('is_canceled == 0').pivot_table(index=['arrival_date_year', 'arrival_date_month'],
                                   values='lead_time',
                                   aggfunc='count').reset_index()

# Заменим в столбце месяца тип object на int с помощью словаря
data_attendance['arrival_date_month'] = data_attendance['arrival_date_month'].map(month_list)
data_attendance = data_attendance.sort_values(by=['arrival_date_year', 'arrival_date_month']).reset_index(drop=True)

# Переведём тип данных в DT для временных рядов
data_attendance['date'] = pd.to_datetime(data_attendance.arrival_date_year.astype(str) + '/' + data_attendance.arrival_date_month.astype(str) + '/01')

# Удалим ненужные столбцы
del data_attendance['arrival_date_year']
del data_attendance['arrival_date_month']

# Установим индекс на дату
data_attendance = data_attendance.set_index('date')

# Нарисует наглядный график посещаемости
data_attendance.plot(grid=True, style='o-', figsize=(15,8))
plt.show()

# Выведем на экран получившуюся картину
display(data_attendance)
```

### Шаг 3. Формулировка ML-задачи на основе бизнес-задачи

Основная **бизнес-метрика** для любой сети отелей **— её прибыль**. Прибыль отеля — это разница между стоимостью номера за все ночи и затраты на обслуживание: как при подготовке номера, так и при проживании постояльца.
В отеле есть несколько типов номеров. В зависимости от типа номера назначается стоимость за одну ночь. Есть также затраты на уборку. Если клиент снял номер надолго, то убираются каждые два дня.

**Стоимость номеров отеля:**

1. Категория **A**: за ночь — 1 000, разовое обслуживание — 400;
2. Категория **B**: за ночь — 800, разовое обслуживание — 350;
3. Категория **C**: за ночь — 600, разовое обслуживание — 350;
4. Категория **D**: за ночь — 550, разовое обслуживание — 150;
5. Категория **E**: за ночь — 500, разовое обслуживание — 150;
6. Категория **F**: за ночь — 450, разовое обслуживание — 150;
7. Категория **G**: за ночь — 350, разовое обслуживание — 150.

В ценовой политике отеля **используются сезонные коэффициенты: весной и осенью цены повышаются на 20%, летом — на 40%.**
Убытки отеля в случае отмены брони номера — это стоимость одной уборки и одной ночи с учётом сезонного коэффициента.
На разработку системы прогнозирования **заложен бюджет — 400 000**. При этом необходимо учесть, что **внедрение модели должно окупиться за тестовый период**. Затраты на разработку должны быть меньше той выручки, которую система принесёт компании.

```python
# Оценим прибыль бизнеса без внедрения депозитов

# Создадим DF с информацией о типе номера и затрат
rooms = pd.DataFrame(data=[['A', 1000, 400], ['B', 800, 350], ['C', 600, 350], ['D', 550, 150], ['E', 500, 150], ['F', 450, 150], ['G', 350, 150]], columns=['reserved_room_type', 'night_cost', 'maintenance_cost'])
display(rooms)

# Создадим DF с информацией о месячном проценте
percent = pd.DataFrame(data=[['January', 1.0], ['February', 1.0], ['March', 1.2], ['April', 1.2], ['May', 1.2], ['June', 1.4], ['July', 1.4], ['August', 1.4], ['September', 1.2], ['October', 1.2], ['November', 1.2], ['December', 1.0]], columns=['arrival_date_month', 'precent'])
display(percent)

# Объеденим все в один DF
data_cost = data.merge(rooms, how='left', on='reserved_room_type')
data_cost = data_cost.merge(percent, how='left', on='arrival_date_month')
# Добавим столбец с кол-во уборки вторая часть сложения - обязательная уборка номера перед заездом (она добавляется только в нечётное кол-во дней)
data_cost['cleaning_quantity'] = (data_cost['total_nights'] // 2) + (data_cost['total_nights'] % 2)
display(data_cost)
```

```python
# Напишем функцию для определения profit (наверх, ко всем функциям переностить не стал, тк она "локальная")
def turnover(row):
    if row['is_canceled'] == 0:
        profit = row['total_nights'] * row['night_cost'] * row['precent'] - (row['maintenance_cost'] * row['cleaning_quantity'])
    else:
        profit = -row['maintenance_cost']
    return int(profit)

data_cost['profit'] = data_cost.apply(turnover, axis=1)
display(data_cost)
data_cost_2017 = data_cost.loc[data_cost['arrival_date_year'] == 2017]
print()
print('+---------------------------------------------+')
print('| Оборот компании без внедрения =', '{:,.0f}'.format(data_cost.loc[data_cost['arrival_date_year'] == 2017, 'profit'].sum()) + '₽ |')
print('+---------------------------------------------+')
```

##### Вывод

- Были созданы новые DF - некие сводки информации по типу номера, месячные проценты
- Также новый столбец по подсчёту кол-ва уборок: 1 раз до приезда и далее через 2 дня
- Новый столбец по расчёту прибыли с каждого клиента
- Оборот компании без внедрения системы прогнозирования за 2017 год составляет 49 миллионов рублей.

```python
# Разобьём общий df по первоначальным признакам, иначе у нас будет неравное кол-во признаков.
# Также пробовал просто разбивать рандомно на test и train результат показывал лучше, но мы пойдём по заданию,
# хотя в другом случае, я бы сделал обычный merge+train_test_split, чтобы увеличить показатели модели.
train = data[:65229].reset_index(drop=True)
test = data[65229:].reset_index(drop=True)
display(train.head(3), test.head(3))
```

```python
# Разделим выборку
features_train, target_train = splitter(data = train, target = 'is_canceled')
features_test, target_test = splitter(data = test, target = 'is_canceled')
print(features_train.shape, target_train.shape, features_test.shape, target_test.shape)
```

```python
# Маштабируем признаки

# Я решил не маштабировать lead_time, previous_cancellations и required_car_parking_spaces, тем самым выделив их и предав
# более сильное значение для модели, нежели остальные
numeric=['stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies', 'previous_bookings_not_canceled', 'booking_changes', 'days_in_waiting_list', 'total_of_special_requests']

scaler = StandardScaler().fit(features_train[numeric])

features_train[numeric] = scaler.transform(features_train[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])

pd.options.mode.chained_assignment = None

display(features_train.describe(), features_test.describe())
```

#### Обучение моделей и дисбаланс

Итак, здесь мы будем использовать class_weight с тремя популярными алгоритмами: дерево решений, логистическая регрессия и случайный лес. Затем выявим победителя и посмотрим матрицу ошибок лучшей модели. Метрикой для оценивания у нас будет recall, ведь она показывает какую долю объектов положительного класса из всех объектов положительного класса нашел алгоритм, что нам и необходимо знать. Простым языком, к примеру, мы смогли обноружить 67 из 100 представленных единиц, тогда наш recall будет равен 67%. Однако хочется видеть полноту картины, поэтому добавим метрику f1 и roc_auc в конце.

```python
%%time

# Дерево решений
final_scores_dtc_cw = []
best_model_dtc_cw = None
best_result_dtc_cw = 0
best_mss_dtc_cw = 0
for depth in range(4, 26):
    print(round((depth - 4) / 21 * 100, 2), "%")
    for mss in range(2, 20):
        model_dtc_cw = DecisionTreeClassifier(criterion='entropy',
                                           splitter = 'random',
                                           random_state=12345,
                                           max_depth = depth,
                                           min_samples_split=mss,
                                           class_weight='balanced')
        scores = cross_val_score(model_dtc_cw,features_train, target_train, cv=5 , scoring='recall')
        result_dtc_cw = sum(scores)/len(scores)
        final_scores_dtc_cw.append(result_dtc_cw)
        if result_dtc_cw > best_result_dtc_cw:
            best_mss_dtc_cw = mss
            best_depth_dtc_cw = depth
            best_model_dtc_cw = model_dtc_cw
            best_result_dtc_cw = result_dtc_cw


print('------------------')
print('Лучший результат:', best_result_dtc_cw,
      '\nЛучшая глубина', best_depth_dtc_cw,
      '\nЛучшее mss', best_mss_dtc_cw,
      '\nЛучшая модель:', best_model_dtc_cw)
```

```python
%%time

# Логистическая регрессия
final_scores_lr_cw = []
best_model_lr_cw = None
best_result_lr_cw = 0
for penalty in ['l1','l2']:
    for c in [0.001, 0.01, 0.1, 1, 10, 100, 1000]:
        model_lr_cw = LogisticRegression(random_state=12345, solver='liblinear', class_weight='balanced', C = c, penalty = penalty)
        scores = cross_val_score(model_lr_cw,features_train, target_train, cv=5 , scoring='recall')
        result_lr_cw = sum(scores)/len(scores)
        final_scores_lr_cw.append(result_lr_cw)
        if result_lr_cw > best_result_lr_cw:
            best_model_lr_cw = model_lr_cw
            best_result_lr_cw = result_lr_cw
            
print('Лучший результат:', best_result_lr_cw,
      '\nЛучшая модель:', best_model_lr_cw)
```

```python
%%time

# Случайный лес
final_scores_rfc_cw = []
best_model_rfc_cw = None
best_est_rfc_cw = 0
best_depth_rfc_cw = 0
best_result_rfc_cw = 0
best_mss_rfc_cw = 0
for est in range(10, 60, 10):
    for depth in range(2, 16):
        for mss in range(2, 16):
            model_rfc_cw = RandomForestClassifier(random_state=12345,
                                                  n_estimators=est,
                                                  max_depth=depth,
                                                  criterion='entropy',
                                                  min_samples_split=mss,
                                                  class_weight='balanced')
            scores = cross_val_score(model_rfc_cw, features_train, target_train, cv=5 , scoring='recall')
            result_rfc_cw = sum(scores)/len(scores)
            final_scores_rfc_cw.append(result_rfc_cw)
            if result_rfc_cw > best_result_rfc_cw:
                best_result_rfc_cw = result_rfc_cw
                best_depth_rfc_cw = depth
                best_est_rfc_cw = est
                best_model_rfc_cw = model_rfc_cw
                best_mss_rfc_cw = mss
print('----------------')
print('Лучший результат', best_result_rfc_cw,
      '\nЛучшее кол-во деревьев', best_est_rfc_cw,
      '\nЛучшая глубина', best_depth_rfc_cw,
      '\nЛучшая модель', best_model_rfc_cw)
```

```python
# Произведём расчёты f1
scores = cross_val_score(best_model_dtc_cw, features_train, target_train, cv=5 , scoring='f1')
f1_dtc = sum(scores)/len(scores)
scores = cross_val_score(best_model_lr_cw, features_train, target_train, cv=5 , scoring='f1')
f1_lr = sum(scores)/len(scores)
scores = cross_val_score(best_model_rfc_cw, features_train, target_train, cv=5 , scoring='f1')
f1_rfc = sum(scores)/len(scores)
```

```python
# Произведём расчёты roc_auc
scores = cross_val_score(best_model_dtc_cw, features_train, target_train, cv=5 , scoring='roc_auc')
roc_auc_dtc = sum(scores)/len(scores)
scores = cross_val_score(best_model_lr_cw, features_train, target_train, cv=5 , scoring='roc_auc')
roc_auc_lr = sum(scores)/len(scores)
scores = cross_val_score(best_model_rfc_cw, features_train, target_train, cv=5 , scoring='roc_auc')
roc_auc_rfc = sum(scores)/len(scores)
```

```python
# Создадим функцию, которая будет отвечать за подсчёт прибыли с учётом предсказаний
def turnover_predict(row):
    if row['is_canceled'] == 0:
        profit = row['total_nights'] * row['night_cost'] * row['precent'] - (row['maintenance_cost'] * row['cleaning_quantity'])
    elif (row['is_canceled'] == 1) and (row['predict'] == 0):
        profit = -row['maintenance_cost']
    elif (row['is_canceled'] == 1) and (row['predict'] == 1):
        profit = row['night_cost']*0.8*row['precent'] + row['maintenance_cost']
    return int(profit)

data_cost_2017['predict_profit'] = data_cost_2017.apply(turnover_predict, axis=1)
display(data_cost_2017)
print()
print('+----------------------------------------------------+')
print('| Оборот компании с внедрением системы =', '{:,.0f}'.format(data_cost_2017['predict_profit'].sum()) + '₽ |')
print('+----------------------------------------------------+')
print()
print('+---------------------------------------------+')
print('| Разница между двумя оборотами =', '{:,.0f}'.format(data_cost_2017['predict_profit'].sum() - data_cost.loc[data_cost['arrival_date_year'] == 2017, 'profit'].sum()) + '₽ |')
print('+---------------------------------------------+')
```
